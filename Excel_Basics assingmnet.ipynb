Q1. Explain the difference between Relative, Absolute and Mixed Cell Referencing.

Answer: Relative Cell Reference (A1):
Changes automatically when a formula is copied to another cell.
Both row and column references change.

Example:
=A1+B1

Absolute Cell Reference ($A$1):
Remains fixed when copied to another cell.
Both row and column references stay constant.

Example:
=A1*$B$1

Mixed Cell Reference ($A1 or A$1):
Either the row or the column remains fixed.

Example:
=$A1*B1

Q2. Write a formula to calculate the total sales of Car and Bicycle only.

Given Data:

Vehicle	Price
Car
500000
Bicycle
10500
Auto-Rickshaw
25000

Answer: =SUM(B2:B3)

Result: 500000 + 10500 = 510500

Total Sales = 510500

Q3. Using the data below, write a formula to calculate the average sales of items priced above 100 but less than 300.

Given Data:

| Item   | Price | Sales |
| ------ | ----- | ----- |
| Item A | 90    | 1000  |
| Item B | 150   | 1200  |
| Item C | 250   | 1500  |


Answer: = AVERAGEIFS(C2:C4,B2:B4,">100",B2:B4,"<300")

Calculation: (1200 + 1500) / 2 = 1350

Average Sales = 1350

Q4. Count how many customer names are recorded.

Answer:

Formula: =COUNTA(B:B)

Explanation:

The COUNTA function counts all non-empty cells in the selected range, giving the total number of customer names recorded.

Q6. Calculate the total sales of Notebooks in the North region only.


Answer:

Formula:=SUMIFS(E:E,C:C,"Notebook",D:D,"North")

Where:

E:E = Total Sales Column
C:C = Product Column
D:D = Region Column

This formula adds only the sales of Notebooks in the North region.

Q8. Insert a line chart showing daily sales trend.

Answer:
